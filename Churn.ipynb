```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set(style= 'darkgrid')
```

```python
data = pd.read_csv('expresso_cleaned.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
num = data.select_dtypes(include= ['number'])
cat = data.select_dtypes(exclude = ['number'])
print('numerical columns:\n')
display(num.head())
print('n\categorical columns:')
display(cat.head())
```

```python
#to know the columns to scale
cols_to_scale = []

ds = data.copy()

for i in num.columns:
    if ds[i].describe()[2] > 1500:
        cols_to_scale.append(i)


cols_to_scale
```

```python

# SCALE THE NUMERICAL COLUMNS
from sklearn.preprocessing import StandardScaler
import joblib

scaled = {}

for i in cols_to_scale[:]:
    scale = StandardScaler()
    ds[i] = scale.fit_transform(ds[[i]])
    joblib.dump(scale, open(f'{i}_scaler.pkl', 'wb'))
    scaled[i + '_scaler']=scale

scaled
```

```python



# to encode categorical datas
from sklearn.preprocessing import LabelEncoder

encoded = {}
for i in cat.columns:
    encode = LabelEncoder()
    ds[i] = encode.fit_transform(ds[i])
    joblib.dump(encode, open(f'{i}_encoder.pkl', 'wb'))
    encoded[i + 'encoder'] = encode

encoded
```

```python
selectedcol = ['DATA_VOLUME', 'ON_NET', 'REGULARITY', 'REVENUE','FREQUENCE', 'MONTANT', 'FREQUENCE_RECH', 'CHURN']

newData = ds[selectedcol]
newData.head()
```

```python
#Split into Train and Test


from sklearn.model_selection import train_test_split

x = newData.drop('CHURN', axis = 1)
y = newData.CHURN


xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, stratify = y)
print(f'xtrain: {xtrain.shape}')
print(f'xtest: {xtest.shape}')
print(f'ytrain: {ytrain.shape}')
print(f'ytest: {ytest.shape}')
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import classification_report
prediction = model.predict(xtrain)
print(classification_report(ytrain, prediction))
```

```python
#fit the algorithm on the data it has never seen before
from sklearn.metrics import classification_report
prediction = model.predict(xtest)
print(classification_report(ytest, prediction))
```

```python

#To export the model

joblib.dump(model, open('Loandmodel.pk1', 'wb'))
```
